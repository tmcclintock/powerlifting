```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
import powerlifting
from powerlifting import wilks, ipf
```

```python
sex = "M"
equipment = "Raw"
event = "SBD"
w = wilks.wilks(sex)
i = ipf.ipf(sex, equipment, event)
```

```python
bw = 200 #lbs
t = np.linspace(500, 2000)
plt.plot(t, w.score(bw, t, in_pounds=True))
plt.plot(t, i.score(bw, t, in_pounds=True))
```

```python
bw = np.linspace(50, 600) #lbs
t = np.linspace(500, 2000)
bws, ts = np.meshgrid(bw, t)
wsc = w.score(bws, ts, in_pounds=True)
isc = i.score(bws, ts, in_pounds=True)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(bws, ts, wsc, color='b', rcount=20, ccount=30)
#ax.plot_wireframe(bws, ts, isc, color='r')
```
